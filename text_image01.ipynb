```python
# 1. Install necessary libraries (This takes about 1-2 mins)
!pip install -q diffusers transformers accelerate gradio torch

import gradio as gr
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# 2. Setup the Model
# Using Stable Diffusion v1.5 for a good balance of speed and quality
model_id = "runwayml/stable-diffusion-v1-5"

# Detect if GPU is available (Colab: Edit -> Notebook Settings -> T4 GPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

print(f"Loading model on {device}...")
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
    use_safetensors=True
)
pipe = pipe.to(device)

# 3. Define the Prediction Function
def generate_image(prompt, negative_prompt, steps, guidance):
    # This function is what Gradio calls when you click 'Generate'
    image = pipe(
        prompt=prompt,
        negative_prompt=negative_prompt,
        num_inference_steps=int(steps),
        guidance_scale=guidance
    ).images[0]
    return image

# 4. Build the Gradio Interface
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("## 🎨 Colab Text-to-Image Studio")
    gr.Markdown("Enter a prompt below and click generate. If you're on a Free Colab tier, ensure GPU is enabled for speed.")

    with gr.Row():
        with gr.Column():
            prompt = gr.Textbox(label="Prompt", placeholder="A steampunk owl wearing glasses, cinematic lighting...")
            neg_prompt = gr.Textbox(label="Negative Prompt", placeholder="blurry, distorted, low quality")

            with gr.Accordion("Advanced Settings", open=False):
                steps = gr.Slider(minimum=1, maximum=50, value=25, step=1, label="Sampling Steps")
                guidance = gr.Slider(minimum=1, maximum=20, value=7.5, step=0.5, label="Guidance Scale")

            generate_btn = gr.Button("Generate Image", variant="primary")

        with gr.Column():
            output_image = gr.Image(label="Generated Result")

    generate_btn.click(
        fn=generate_image,
        inputs=[prompt, neg_prompt, steps, guidance],
        outputs=output_image
    )

# 5. Launch the App
# share=True creates a public URL (gradio.live) so you can share the link
demo.launch(share=True, debug=True)
```
