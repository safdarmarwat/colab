```python
# Install libraries
!pip install -q gradio requests

import gradio as gr
import requests
import os

# Set your key here
API_KEY = "4c11b09b498a4b55b0463328261204"

def get_weather_data(city):
    if not city: return "Enter a city name."

    # WeatherAPI URL for current weather
    url = f"http://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}&aqi=no"

    try:
        response = requests.get(url)
        data = response.json()

        if "error" in data:
            return f"❌ Error: {data['error']['message']}"

        # Extracting data
        location = data['location']['name']
        region = data['location']['region']
        temp_c = data['current']['temp_c']
        condition = data['current']['condition']['text']
        icon_url = "https:" + data['current']['condition']['icon']
        humidity = data['current']['humidity']
        wind_kph = data['current']['wind_kph']

        return f"""
        ### 📍 {location}, {region}
        ![weather icon]({icon_url})
        **Condition:** {condition}
        **Temperature:** {temp_c}°C
        **Humidity:** {humidity}%
        **Wind Speed:** {wind_kph} km/h
        """
    except Exception as e:
        return f"⚠️ System Error: {str(e)}"

# Gradio UI
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🌦️ WeatherAPI Dashboard")
    city_input = gr.Textbox(label="Search City", placeholder="e.g. Paris, Mumbai, New York")
    out = gr.Markdown()
    city_input.submit(get_weather_data, city_input, out)

demo.launch()
```
