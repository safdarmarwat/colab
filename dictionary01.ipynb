```python
# 1. Install Library
!pip install groq -q

import os
from groq import Groq
from google.colab import userdata

# 2. Setup Client (Make sure GROQ_API_KEY is in your Colab Secrets)
client = Groq(api_key=userdata.get("GROQ_API_KEY"))

def mento_ai_assistant(skill, user_problem, mode="overview"):
    """
    skill: e.g., 'Python Dictionaries'
    user_problem: The specific issue they are stuck on.
    mode: 'overview' for logic, 'deep_dive' for technical detail.
    """

    # Custom System Instruction based on the Mode
    if mode == "overview":
        system_msg = f"You are a Senior Architect. Focus on the logic and 'why' of using {skill}. Do not write all the code, help the user think through the problem."
    else:
        system_msg = f"You are a Lead Developer. Provide a deep dive into {skill}. Review the user's code for efficiency, $O(n)$ complexity, and professional best practices."

    completion = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user", "content": f"I am working on this problem: {user_problem}"}
        ],
        temperature=0.7,
    )

    print(f"\n--- {mode.upper()} MODE ---")
    print(completion.choices[0].message.content)

# --- TEST IT OUT ---

# Example Problem: User wants to build a simple frequency counter for words in a book
my_skill = "Python Dictionaries"
my_problem = "I want to count how many times each word appears in a text file, but my code is slow and messy."

# Try Overview
mento_ai_assistant(my_skill, my_problem, mode="overview")

# Try Deep Dive
# mento_ai_assistant(my_skill, my_problem, mode="deep_dive")
```
