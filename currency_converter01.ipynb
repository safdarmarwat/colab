```python
# Currency Converter (Basic Version)

def convert_currency(amount, from_currency, to_currency):
    # Example fixed exchange rates (relative to USD)
    rates = {
        "USD": 1.0,
        "PKR": 278.0,
        "EUR": 0.92,
        "GBP": 0.78,
        "INR": 83.0,
        "AED": 3.67
    }

    # Convert to USD first
    usd_amount = amount / rates[from_currency]

    # Convert USD to target currency
    converted = usd_amount * rates[to_currency]

    return converted


while True:
    print("\n--- Currency Converter ---")
    print("Available currencies: USD, PKR, EUR, GBP, INR, AED")

    amount = float(input("Enter amount: "))
    from_currency = input("From currency: ").upper()
    to_currency = input("To currency: ").upper()

    if from_currency in ["USD", "PKR", "EUR", "GBP", "INR", "AED"] and \
       to_currency in ["USD", "PKR", "EUR", "GBP", "INR", "AED"]:

        result = convert_currency(amount, from_currency, to_currency)
        print(f"Converted Amount: {result:.2f} {to_currency}")
    else:
        print("❌ Invalid currency code!")

    choice = input("Convert again? (y/n): ")
    if choice.lower() != 'y':
        print("Stopped.")
        break
```
