```python
import json
import os
from datetime import date

FILE_NAME = "habits.json"

# Load habits from file
def load_data():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return json.load(f)
    return {}

# Save habits to file
def save_data(data):
    with open(FILE_NAME, "w") as f:
        json.dump(data, f)

# Add new habit
def add_habit(data):
    habit = input("Enter new habit: ")
    if habit not in data:
        data[habit] = []
        print("Habit added!")
    else:
        print("Habit already exists!")

# Mark habit as done
def mark_done(data):
    habit = input("Enter habit name: ")
    if habit in data:
        today = str(date.today())
        if today not in data[habit]:
            data[habit].append(today)
            print("Marked as done ✅")
        else:
            print("Already completed today!")
    else:
        print("Habit not found!")

# View habits
def view_habits(data):
    if not data:
        print("No habits yet.")
        return

    print("\nYour Habits:")
    for habit, days in data.items():
        print(f"- {habit} (Completed {len(days)} days)")

# Main program
data = load_data()

while True:
    print("\n--- Habit Tracker ---")
    print("1. Add Habit")
    print("2. Mark Done")
    print("3. View Habits")
    print("4. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        add_habit(data)
    elif choice == "2":
        mark_done(data)
    elif choice == "3":
        view_habits(data)
    elif choice == "4":
        save_data(data)
        print("Data saved. Goodbye!")
        break
    else:
        print("Invalid choice!")
```
